# Let's make some prediction models

```python
df.columns
```

```python
x = df[[ 'Region', 'HDI',
       'GDP_PerCapita', 'Beer_PerCapita', 'Spirit_PerCapita', 'Wine_PerCapita',
       ]]
y = df['HappinessScore']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=38)
```

Converting String into integers

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
x_train['Region'] = label_encoder.fit_transform(x_train['Region'])
x_test['Region'] = label_encoder.fit_transform(x_test['Region'])
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
linear.score(x_test,y_test)
```

```python
y_test_pred = linear.predict(x_test)
plt.plot(y_test,y_test_pred)
plt.xlabel('Actual Value')
plt.ylabel('predicted value')
```

Ensemble Techniques

```python
from sklearn.ensemble import GradientBoostingRegressor
boost = GradientBoostingRegressor()
boost.fit(x_train,y_train)
boost.score(x_test,y_test)
```

```python
y_test_pred = boost.predict(x_test)
plt.plot(y_test,y_test_pred,color='red')
plt.xlabel('Actual Value')
plt.ylabel('predicted value')
```
